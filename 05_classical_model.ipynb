```python
"""

Classical Model / Modelo Clássico

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""
Labor Market / Mercado de Trabalho

L^s = L_0^s + aw

L^d = L_0^d - bw

"""
w = np.linspace(0,100, 50)
L_S, L_D = np.zeros(50), np.zeros(50)

for i in range(0, 50):
    L_S[i], L_D[i] = 1.5*w[i], 100 - 0.7*w[i]
    
fig, ax = plt.subplots()
ax.plot(L_S,w,'b')
ax.plot(L_D,w,'r')
ax.set_ylabel('w')
ax.set_xlabel('L_S, L_D')
plt.show()

L = L_S[np.argmin((L_S-L_D)**2)]
print("L^* =", L, ",w^*=", w[np.argmin((L_S-L_D)**2)]  )
```

```python
"""
Funds Market / Mercado de Fundos Emprestáveis

S = S_0 + cr

I = I_0 - dr

"""
r = np.linspace(0,1, 50)
S, I = np.zeros(50), np.zeros(50)

for i in range(0, 50):
    S[i], I[i] = 25 + 5*r[i], 30- 5*r[i]
    
fig, ax = plt.subplots()
ax.plot(S,r,'b')
ax.plot(I,r,'r')
ax.set_ylabel('r')
ax.set_xlabel('S,I')
plt.show()

K = I[np.argmin((S-I)**2)]
print("K^* =", K, ",r^*=", r[np.argmin((S-I)**2)]  )
```

```python
"""
Aggregate Supply and Demand / Oferta e Demanda Agregada

AS = alpha*log(K) + (1-alpha)*log(L)

AD = log(M) + log(V) - log(P)

"""

logP = np.linspace(72.5,80, 50)
AS, AD = np.zeros(50), np.zeros(50)

for i in range(0, 50):
    AS[i], AD[i] = 0.3*np.log(K) + (1-0.3)*np.log(L), 80 - logP[i]
    
fig, ax = plt.subplots()
ax.plot(AS,logP,'b')
ax.plot(AD,logP,'r')
ax.set_ylabel('log(P)')
ax.set_xlabel('log(AS),log(AD)')
plt.show()

logY = 0.3*np.log(K) + (1-0.3)*np.log(L)
print("logY =", logY, ",P^*=", logP[np.argmin((AS-AD)**2)]  )

```
