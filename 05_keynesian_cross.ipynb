```python
"""

Keynesian Cross / Cruz Keynesiana

Author / Autor: imoripaolo

"""

import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
"""
Aggregate Supply / Oferta Agregada: AS = Y

Aggregate Demand / Demanda Agregada: AD = AD_0 + c_1(Y-T) + G

Market Clearing / Equilíbrio de Mercado: AS = AD

"""

Y = np.linspace(0,200, 50)
AS, AD = np.zeros(50), np.zeros(50)
T, G = 50, 70

for i in range(0, 50):
    AS[i], AD[i] = Y[i], 20 + 0.3*(Y[i]-T) + G

fig, ax = plt.subplots()
ax.plot(AS,Y,'b')
ax.plot(Y,AD,'r')
ax.set_ylabel('AD')
ax.set_xlabel('AS')
plt.show()

print("Y^* =", AD[np.argmin((AS-AD)**2)])
```

```python
"""
Fiscal Shock / Choque Fiscal: dG = 30

"""

AD1, G1 = np.zeros(50), G+30
for i in range(0, 50):
    AD1[i] = 20 + 0.3*(Y[i]-T) + G1
    
fig, ax = plt.subplots()
ax.plot(AS,Y,'b')
ax.plot(Y,AD,'b')
ax.plot(Y,AD1,'r')
ax.set_ylabel('AD')
ax.set_xlabel('AS')
plt.show()

print("Y^* =", AD1[np.argmin((AS-AD1)**2)], "dY=", AD1[np.argmin((AS-AD1)**2)]- AD[np.argmin((AS-AD)**2)])
```
